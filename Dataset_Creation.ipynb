# Dataset Creation

```python
import pandas as pd
import numpy as np
import random

from sklearn.model_selection import train_test_split
```

```python
# Création des listes aléatoires pour créer les lignes de notre Dataframe
ips_list = [] 
sex_viol_list = [] 
miea_list = [] 
drp_list = []
att_list = []  
social_inf_list = []  
sex_pref_list = []  
para_list = [] 
child_emo_list = []  
impul_list = [] 
outlaw_list = [] 
addict_list = []

ips_value = ['Low', 'Medium', 'High']
sex_viol_val = ['Yes', 'No']
miea_val = ['Yes', 'No']
drp_val = ['Yes', 'No']
att_val = ['Yes', 'No']
social_inf_val = ['Positive', 'Negative'] 
sex_pref_val = ['Young Puberty', 'Adult']
para_val = ['Yes', 'No']
child_emo_val = ['Yes', 'No']
impul_val = ['Yes', 'No']
outlaw_val = ['Yes', 'No']
addict_val = ['Yes', 'No']

ips_index = np.random.randint(3, size=700)
sex_viol_index = np.random.randint(2, size=700)
miea_index = np.random.randint(2, size=700)
drp_index = np.random.randint(2, size=700)
att_index = np.random.randint(2, size=700)
social_inf_index = np.random.randint(2, size=700)
sex_pref_index = np.random.randint(2, size=700)
para_index = np.random.randint(2, size=700)
child_emo_index = np.random.randint(2, size=700)
impul_index = np.random.randint(2, size=700)
outlaw_index = np.random.randint(2, size=700)
addict_index = np.random.randint(2, size=700)

for i in ips_index:
    ips_list.append(ips_value[i])
for i in sex_viol_index:
    sex_viol_list.append(sex_viol_val[i])
for i in miea_index:
    miea_list.append(miea_val[i])
for i in drp_index:
    drp_list.append(drp_val[i])
for i in att_index:
    att_list.append(att_val[i])
for i in social_inf_index:
    social_inf_list.append(social_inf_val[i])
for i in sex_pref_index:
    sex_pref_list.append(sex_pref_val[i])
for i in para_index:
    para_list.append(para_val[i])
for i in child_emo_index:
    child_emo_list.append(child_emo_val[i])
for i in impul_index:
    impul_list.append(impul_val[i])
for i in outlaw_index:
    outlaw_list.append(outlaw_val[i])
for i in addict_index:
    addict_list.append(addict_val[i])

#Id Part
nbre_id = random.sample(range(1000, 2000), 700)
id_list = []
for i in nbre_id:
    res = 'C' + str(i)
    id_list.append(res)
```

```python
dico = {
    "Suject_ID" : id_list,
    "IPS" : ips_list,
   "Sexual_Violence" : sex_viol_list,
    "MIEA" : miea_list,
    "DRP" : drp_list,
    "Attitude" : att_list,
    "Social_Influence" : social_inf_list,
    "Sexual_Preference" : sex_pref_list,
    "Paraphilia" : para_list,
    "Child_Emotion" : child_emo_list,
    "Impulsiveness" : impul_list,
    "Outlaw" : outlaw_list,
    "Addict" : addict_list
}

#Create a DataFrame
dataframe = pd.DataFrame(dico)
dataframe.head()
```
