# Predict a state of recurrence.

```python
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, GridSearchCV, learning_curve
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
import sklearn.metrics

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.simplefilter('ignore')
```

```python
data = pd.read_csv('Psychological_Data.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
# Transform Data
yes_no_cols = ['MIEA', 'DRP', 'Attitude','Paraphilia', 'Child_Emotion' ,'Impulsiveness' ,'Outlaw' ,'Addict', 'Sexual_Violence']
for i in yes_no_cols:
    data[i] = data[i].map({"Yes" : 1, "No" : 0})
    
data['IPS'] = data['IPS'].map({"Low" : 0, "Medium" : 1, "High" : 2})
data['Social_Influence'] = data['Social_Influence'].map({"Positive" : 1, "Negative" : 0})
data['Sexual_Preference'] = data['Sexual_Preference'].map({"Young Puberty" : 1, "Adult" : 0})

data.head()
```

```python
# Creation of the Recidivist Column
cols_addition = 0
for i in list(data.columns[1:]):
    cols_addition = cols_addition + data[i]

data['Recidivist'] = cols_addition
data.head()
```

```python
reci = list(data['Recidivist'])
new_rec_val = []
for i in reci:
    if i > 5:
        new_rec_val.append(1)
    else:
        new_rec_val.append(0)
data['Recidivist'] = new_rec_val
data.head()
```

```python
#Séparation de nos données Train et Test
train = data.iloc[0 : 434, :]
test = data.iloc[434 : 700, 0:13]

convert_test = test.to_csv("test_data.csv", encoding="utf-8",index=False)
convert_train = train.to_csv("train_data.csv", encoding="utf-8", index=False)
```

```python
def Evaluate_Model(model, x, y):
    cv = StratifiedKFold(5)
    score = cross_val_score(model, x, y, cv = cv)
    return round(np.mean(score)*100, 2)
```

```python
X = train_data.drop('Recidivist', axis = 1)
y = train_data['Recidivist']
```

```python
rl_params = {
    "penalty" : ['l1', 'l2', 'elasticnet', None],
    "dual" : [True, False],
    "fit_intercept" : [True, False],
    "solver" : ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    "max_iter" : [200]
}
rl_grid = GridSearchCV(LogisticRegression(), rl_params, cv = 5)
rl_grid.fit(X, y)
rl_model = rl_grid.best_estimator_
print(("Logistic Regression Score : {}%").format(Evaluate_Model(rl_model, X, y)))
```

```python
dt_params = {
    "criterion" : ['gini', 'entropy'],
    "splitter" : ['best', 'random'],
    "max_depth" : [5, 6, 7, 8, 9, 10]
}
dt_grid = GridSearchCV(DecisionTreeClassifier(), dt_params, cv = 5)
dt_grid.fit(X, y)
dt_model = dt_grid.best_estimator_
print(("Decision Tree Score : {}%").format(Evaluate_Model(dt_model, X, y)))
```

```python
N, train_score, val_score = learning_curve(dt_model, X, y, train_sizes=np.linspace(0.1, 1.0, 10), cv=5)
plt.figure(figsize=(20, 8))
plt.plot(N, train_score.mean(axis = 1), label = "Train")
plt.plot(N, val_score.mean(axis = 1), label = "Validation")
plt.title("Decision Tree Evolution")
plt.legend()
```

```python
rf_params = {
    "bootstrap" : [True],
    "n_estimators" : [100, 200, 300],
    "criterion" : ['gini', 'entropy'],
    "max_depth" : [10, 20, 30, 40, 50, 80, 90, 100, 110, 120],
    "max_features": [2, 3]
}
rf_grid = GridSearchCV(RandomForestClassifier(), rf_params, cv=3, n_jobs = -1, verbose = 2)
rf_grid.fit(X, y)
rf_model = rf_grid.best_estimator_
print(("Random Forest Score : {}%").format(Evaluate_Model(rf_model, X, y)))
```

```python
knn_params = {
    "n_neighbors" : list(np.arange(1, 20)),
    "weights" : ['uniform', 'distance'],
    "algorithm" : ['auto', 'ball_tree', 'kd_tree', 'brute'],
    "metric" : ['euclidean', 'manhattan', 'minkowski']
}

knn_grid = GridSearchCV(KNeighborsClassifier(), knn_params, cv = 5)
knn_grid.fit(X, y)
knn_model = knn_grid.best_estimator_
print(("KNN Score : {}%").format(Evaluate_Model(rf_model, X, y)))
```
